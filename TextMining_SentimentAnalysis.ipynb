## Tokenizing Comments

```python
#make one comment to a list of words
from nltk.tokenize import word_tokenize
```

```python
comment['CommentsTokenized'] = comment['Comments'].apply(word_tokenize)
```

```python
comment
```

```python
#Creating a new dataframe
new_comment=pd.DataFrame()
new_comment=comment.drop(columns=["CommentsTokenized","Comments"])
new_comment
```

## Snowball Stemmer

```python
from nltk.stem.snowball import SnowballStemmer
stemmer1 = SnowballStemmer("english")
new_comment['CommentsTokenizedStemmed'] = comment['CommentsTokenized'].apply(lambda x: [stemmer1.stem(y) for y in x]) # Stem every word.
```

```python
new_comment
```

```python
new_comment['CommentsTokenizedStemmed'] = new_comment['CommentsTokenizedStemmed'].apply(lambda x: " ".join(x))
```

```python
new_comment
```

```python
new_comment1=comment.drop(columns=["CommentsTokenized","Comments"])
new_comment1
```

## Porter Stemmer

```python
from nltk.stem.porter import *
stemmer2 = PorterStemmer()
new_comment1['CommentsTokenizedStemmed'] = comment['CommentsTokenized'].apply(lambda x: [stemmer2.stem(y) for y in x]) # Stem every word.
```

```python
new_comment1
```

## TF-IDF matrix

```python
nltk.download('stopwords')
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
count_vect=CountVectorizer(stop_words=stop_words,lowercase=False)
TD_counts=count_vect.fit_transform(new_comment.CommentsTokenizedStemmed)
DF_TD_Counts=pd.DataFrame(TD_counts.toarray())
```

```python
DF_TD_Counts
```

## Using TF-IDF to find important variables

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
tfidf_count = tfidf_transformer.fit_transform(DF_TD_Counts)
```

```python
DF_tfidf=pd.DataFrame(tfidf_count.toarray())
DF_tfidf
```

## Concating data

```python
data=pd.concat([DF_tfidf,customers],axis=1)
```

```python
data
```

## Dummies

```python
feature=data.drop(columns=['TARGET','ID'])
target=data['TARGET']
```

```python
object_columns=list(feature.select_dtypes(include=['object']).columns)
object_columns
```

```python
dummy_feature=pd.get_dummies(feature,columns=object_columns)
```

```python
dummy_feature
```

```python
from sklearn import preprocessing
dummy_target=preprocessing.LabelEncoder().fit_transform(target)
```

```python
dummy_target
```

## Feature Selection - Filter Method

```python
#Feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
filter_feature = SelectKBest(score_func=chi2, k=15).fit_transform(dummy_feature,dummy_target)
filter_feature.shape
```

## Feature Selection - Wrapper Method

```python
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
knn = KNeighborsClassifier(n_neighbors=4)
sfs1 = SFS(knn, 
           k_features=15, 
           forward=True, 
           floating=False, 
           verbose=2,
           cv=5)

sfs1 = sfs1.fit(dummy_feature,dummy_target)
```

```python
wrapper_feature_name=list(sfs1.k_feature_names_)
```

## Random Forest Model with Feature Selection - Filter Method

```python
from sklearn.model_selection import train_test_split
x_train,y_train,x_target,y_target=train_test_split(filter_feature,dummy_target,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train, x_target)
predict_rf = rf.predict(y_train)
predict_rf
```

```python
from sklearn.metrics import accuracy_score, classification_report
#accuracy_score
print(accuracy_score(y_target,predict_rf))
#see other scores
print(classification_report(y_target,predict_rf))
```

## Random Forest Model with Feature Selection - Wrapper Method

```python
wrapper_feature=pd.DataFrame(dummy_feature[wrapper_feature_name])
```

```python
f_train,f_test,t_train,t_test=train_test_split(wrapper_feature,dummy_target,test_size=0.2)
```

```python
rf1=RandomForestClassifier()
rf1.fit(f_train,t_train)
predict_rf1 = rf1.predict(f_test)
predict_rf1
```

## Prediciton Score

```python
#accuracy_score
print(accuracy_score(t_test,predict_rf1))
#see other scores
print(classification_report(t_test,predict_rf1))
```
